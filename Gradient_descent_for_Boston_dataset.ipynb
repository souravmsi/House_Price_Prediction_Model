```python
'''Importing Libraries'''
import numpy as np
import pandas as pd
df = pd.read_csv('training_boston_x_y_train.csv')
df.head()
```

```python
clmn = df.columns
print(''' Columns :-  ''')
for idx, i in enumerate (clmn):
    print(idx+1, " ", i)
```

```python
df["area_square"]=df[" ZN"]**2
df["room_square"]=df[" RM"]**2
df["crime_square"]=df["# CRIM"]**2
df["tax_square"]=df[" TAX"]**2
df['constant'] = 1
y_train = df[' Y']
del df[' Y']
x_train = df.values
```

```python
print('''Dimensions of x_train and y_train :-''')
print(x_train.shape,", ", y_train.shape)
```

```python
'''cost funtion which will calculate the cost at given points'''
def cost(xt,yt,m):
    M = len(xt)
    m_array_size = len(m)
    total_cost = 0
    for i in range(M):
        y = yt[i]
        x = xt[i]
        total_cost += (1/M) * ((y - (m * x).sum())**2)
    return total_cost
```

```python
'''Step Gradient Function which will calculate the new slope values'''
def step_gradient(xt,yt, learning_rate, m):
    m_slope = np.zeros(18)
    M = len(xt)
    for i in range(M):
        x = xt[i]
        y = yt[i]
        m_slope+=(-2/M)*((y-(m*x).sum())*x)
        
    m -= (learning_rate*m_slope)
    
    return m
```

```python
'''Gradient Descent function'''
def gd(x,y,learning_rate,num_iterations):
    size = 18
    m_array = np.zeros(size)
    for i in range(num_iterations):
        m_array = step_gradient(x,y,learning_rate,m_array)
        print(i, " Cost: ", cost(x,y,m_array))
    return m_array
```

```python
'''This is a Driver Function which will return slope array'''
def run():
    learning_rate = 0.018
    num_iterations = 1000
    m_array = gd(x_train,y_train,learning_rate,num_iterations)
    print(m_array)
    return m_array
```

```python
m_array = run()
```

# Predictions

```python
'''Preprocessing the testing data set'''
test_sample = pd.read_csv("test_boston_x_test.csv",delimiter=',',header=None)
x_test = pd.DataFrame(test_sample)
test_sample.columns = clmn[:13]
x_test["area_square"]=x_test[" ZN"]**2
x_test["room_square"]=x_test[" RM"]**2
x_test["crime_square"]=x_test["# CRIM"]**2
x_test["tax_square"]=x_test[" TAX"]**2
x_test['constant'] = 1
```

```python
'''Calculating the predicted values for the given testing dataset'''
pred = []
for i in range(len(x_test)):
    pred.append((x_test.values[i] * m_array).sum())
```
